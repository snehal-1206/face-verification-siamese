```python
anchor = tf.data.Dataset.list_files(
    os.path.join(ANC_PATH, '*.jpg'),
    shuffle=True
).take(300)

positive = tf.data.Dataset.list_files(
    os.path.join(POS_PATH, '*.jpg'),
    shuffle=True
).take(300)

negative = tf.data.Dataset.list_files(
    os.path.join(NEG_PATH, '*', '*.jpg'),
    shuffle=True
).take(300)

```

```python
def preprocess(file_path):
    byte_img = tf.io.read_file(file_path)
    img = tf.io.decode_jpeg(byte_img, channels=3)
    img = tf.image.resize(img, (100,100))
    img = img / 255.0
    return img
```

```python
positives = tf.data.Dataset.zip((
    anchor, positive, tf.data.Dataset.from_tensor_slices(tf.ones(300))
))

negatives = tf.data.Dataset.zip((
    anchor, negative, tf.data.Dataset.from_tensor_slices(tf.zeros(300))
))

data = positives.concatenate(negatives)
```

```python
def preprocess_twin(img1, img2, label):
    return preprocess(img1), preprocess(img2), label

data = data.map(preprocess_twin)
data = data.cache().shuffle(1024)
```

```python
train_data = data.take(int(len(data)*0.7))
train_data = train_data.batch(16).prefetch(8)

test_data = data.skip(int(len(data)*0.7))
test_data = test_data.batch(16).prefetch(8)
```

```python
def make_embedding():
    inp = Input((100,100,3))
    x = Conv2D(64, (10,10), activation='relu')(inp)
    x = MaxPooling2D()(x)
    x = Conv2D(128, (7,7), activation='relu')(x)
    x = MaxPooling2D()(x)
    x = Conv2D(128, (4,4), activation='relu')(x)
    x = MaxPooling2D()(x)
    x = Conv2D(256, (4,4), activation='relu')(x)
    x = Flatten()(x)
    x = Dense(4096, activation='sigmoid')(x)
    return Model(inp, x)

embedding = make_embedding()
```

```python
class L1Dist(Layer):
    def call(self, x1, x2):
        return tf.math.abs(x1 - x2)
```

```python
def make_siamese_model():
    input_img = Input((100,100,3))
    val_img = Input((100,100,3))

    emb1 = embedding(input_img)
    emb2 = embedding(val_img)

    dist = L1Dist()(emb1, emb2)
    output = Dense(1, activation='sigmoid')(dist)

    return Model([input_img, val_img], output)

siamese_model = make_siamese_model()
```

```python
loss_fn = tf.keras.losses.BinaryCrossentropy()
optimizer = tf.keras.optimizers.Adam(1e-4)
```

```python
@tf.function
def train_step(batch):
    x, y = batch[:2], batch[2]
    with tf.GradientTape() as tape:
        pred = siamese_model(x, training=True)
        loss = loss_fn(y, pred)
    grads = tape.gradient(loss, siamese_model.trainable_variables)
    optimizer.apply_gradients(zip(grads, siamese_model.trainable_variables))
    return loss
```

```python
def train(data, epochs):
    for epoch in range(epochs):
        print(f"\nEpoch {epoch+1}/{epochs}")
        for batch in data:
            train_step(batch)

train(train_data, 50)




```

```python
siamese_model.save('siamese_model.h5')

model = tf.keras.models.load_model(
    'siamese_model.h5',
    custom_objects={'L1Dist': L1Dist}
)
```

```python
def verify(model, det_thresh, ver_thresh):
    results = []
    for img in os.listdir('application_data/verification_images'):
        input_img = preprocess('application_data/input_image/input_image.jpg')
        val_img = preprocess(os.path.join('application_data/verification_images', img))

        pred = model.predict([np.expand_dims(input_img,0),
                              np.expand_dims(val_img,0)])
        results.append(pred)

    detection = np.sum(np.array(results) > det_thresh)
    verification = detection / len(results)
    return verification > ver_thresh
```

```python
cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    frame = frame[120:370,200:450]

    cv2.imshow('Verification', frame)

    if cv2.waitKey(10) & 0xFF == ord('v'):
        cv2.imwrite('application_data/input_image/input_image.jpg', frame)
        print("Verified:", verify(model, 0.8, 0.6))

    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
